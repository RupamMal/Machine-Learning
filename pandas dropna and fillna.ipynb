# Dropping Missing Data using dropna()

```python
data.shape
```

```python
data.dropna()
```

```python
data.dropna(how = 'all')
```

```python
data.dropna(how = 'all').shape
```

```python
data.dropna(axis = 1)
```

```python
data.dropna(axis = 1).shape
```

```python
data.dropna(axis = 1, how = 'all')
```

```python
data.dropna(subset = 'Cabin')
```

```python
data.dropna(subset = 'Cabin').shape
```

```python
data.dropna(subset = 'Embarked')
```

```python
data.dropna(subset = 'Embarked').shape
```

```python
data.dropna(subset = ['Embarked','Cabin'])
```

```python
data.dropna(subset = ['Embarked','Cabin']).shape
```
